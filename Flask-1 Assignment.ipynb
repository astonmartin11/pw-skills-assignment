Flask Framework: 

Flask is a lightweight web framework for Python that is designed to build web applications quickly and with minimal effort. It is based on the WSGI (Web Server Gateway Interface) and the Jinja2 templating engine.

Advantages of Flask:

1)Lightweight & Simple: Minimal setup required, making it easy to start.

2)Flexible: You can customize components as needed.

3)Built-in Development Server: Makes debugging and testing easy.

4)Supports RESTful APIs: Useful for building APIs quickly.

5)Extensible: Can integrate with various extensions (e.g., Flask-SQLAlchemy, Flask- RESTful).

6)Modular Design: You can break applications into smaller modules (Blueprints).

7)Great for Beginners & Small Projects: Easy to learn and use.


# Solution 2

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)
```

App Routing in Flask:

App Routing in Flask refers to mapping URLs to specific functions using the @app.route() decorator.

Why Use App Routes?
    
1)Defines URL Structure: Organizes how users navigate a website.

2)Handles Multiple Pages: Each route links to a different function/page.

3)Dynamic URLs: Routes can accept parameters (e.g., @app.route('/user/<name>')).

4)Improves Code Readability: Separates logic for different pages.

# solution 4

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def company_info():
    return '''<h1>Company Name: ABC Corporation</h1>
              <p>Location: India</p>
              <p>Contact Detail: 999-999-9999</p>'''

@app.route('/welcome')
def welcome():
    return "<h2>Welcome to ABC Corporation</h2>"

if __name__ == '__main__':
    app.run(debug=True)
```

# Solution 5

Function Used:
The url_for() function is used to dynamically build URLs in Flask. It takes the function name and generates a URL for it.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/contact')
def contact():
    return 'Contact Page'

@app.route('/show_urls')
def show_urls():
    return f'''
        <p>Home URL: {url_for('home')}</p>
        <p>About URL: {url_for('about')}</p>
        <p>Contact URL: {url_for('contact')}</p>
    '''

if __name__ == '__main__':
    app.run(debug=True)
```
