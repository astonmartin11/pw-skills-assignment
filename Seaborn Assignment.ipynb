# Solution 1

1)Line Plot (sns.lineplot()):
Used to visualize data trends over time or sequential measurements.

2)Bar Plot (sns.barplot()):
Used to compare the mean or aggregate values of categories.

3)Box Plot (sns.boxplot()):
Used to visualize the distribution of data and detect outliers.

4)Histogram (sns.histplot()):
Used to understand the frequency distribution of a single variable.

5)Pair Plot (sns.pairplot()):
Used to visualize pairwise relationships in a dataset, especially for exploratory data analysis.

# Solution 2

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
fmri = sns.load_dataset("fmri")

# Line plot: signal over timepoint for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time")
plt.show()
```

# Solution 3

```python
# Load dataset
titanic = sns.load_dataset("titanic")

# Box plot for Age vs Pclass
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Pclass")

# Box plot for Fare vs Pclass
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Pclass")

plt.tight_layout()
plt.show()
```

# Solution 4

```python
# Load dataset
diamonds = sns.load_dataset("diamonds")

# Histogram with hue
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", kde=False)
plt.title("Distribution of Diamond Prices by Cut")
plt.show()
```

# Solution 5

```python
# Load dataset
iris = sns.load_dataset("iris")

# Pair plot with species as hue
sns.pairplot(data=iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

# Solution 6

```python
# Load dataset and pivot for heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Heatmap (Passengers per Month/Year)")
plt.show()
```
