In Python, all built-in exceptions inherit from the Exception class, which is the base class for all exceptions. When creating custom exceptions, we derive them from Exception to ensure that they behave like standard exceptions and can be caught using except Exception:. # Solution 4

```python
try:# example of key error
    my_dict = {"name": "Alice"}
    print(my_dict["age"])  # Key does not exist
except KeyError as e:
    print("KeyError:", e)
```

```python
try:# example of index error
    my_list = [1, 2, 3]
    print(my_list[5])  # Index out of range
except IndexError as e:
    print("IndexError:", e)
```
