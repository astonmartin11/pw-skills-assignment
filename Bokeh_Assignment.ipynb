# Solution 3

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 7, 2, 4, 6]

# Create figure
p = figure(title="Customized Plot", x_axis_label="X Axis", y_axis_label="Y Axis")

# Add line with legend
p.line(x, y, legend_label="My Line", line_width=2, line_color="green")

# Customize title
p.title.text_color = "blue"
p.title.text_font_size = "16pt"

# Customize axes
p.xaxis.axis_label_text_color = "red"
p.yaxis.axis_label_text_font_style = "italic"

# Customize legend
p.legend.location = "top_left"
p.legend.label_text_font_size = "12pt"

# Show plot
show(p)
```

# Solution 4

A Bokeh server allows Bokeh apps to have real-time interactivity using Python callbacks.

It allows plots to update dynamically when user input (like sliders, buttons) changes.

 Steps:

Create a Python script with interactive widgets and callbacks.

Run using bokeh serve --show your_script.py

```python
from bokeh.io import curdoc
from bokeh.models import Slider
from bokeh.plotting import figure
from bokeh.layouts import column

slider = Slider(start=0, end=10, value=1, step=0.1, title="Value")
plot = figure()
line = plot.line([1, 2, 3], [1, 2, 3])

def update(attr, old, new):
    line.data_source.data['y'] = [i * slider.value for i in [1, 2, 3]]

slider.on_change('value', update)
curdoc().add_root(column(slider, plot))
```

# Solution 5

```python
from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure

app = Flask(__name__)

@app.route('/')
def index():
    plot = figure(title="Bokeh + Flask")
    plot.circle([1, 2, 3], [4, 6, 5], size=10)

    script, div = components(plot)
    return render_template("index.html", script=script, div=div)

# Template (index.html) must include {{ script|safe }} and {{ div|safe }}
```
