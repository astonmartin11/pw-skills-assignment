# Solution 2

A scatter plot displays values for two numeric variables using dots. It's useful to visualize the relationship or correlation between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data generation
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Scatter plot
plt.scatter(x, y, color='green')
plt.title("Scatter Plot of Random Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

# Solution 3

The subplot() function allows you to draw multiple plots (figures) in a single window, organized in a grid (rows × columns). This is helpful for comparing multiple datasets or visualizations side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Create a 2x2 subplot
plt.figure(figsize=(10, 8))

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='red')
plt.title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple')
plt.title("Line 4")

plt.tight_layout()
plt.show()
```

# Solution 4

A bar plot displays categorical data using rectangular bars. It's used to compare values across different categories. A horizontal bar plot is similar but bars go sideways.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit, color='skyblue')
plt.title("Company Profits - Vertical Bar")
plt.ylabel("Profit ($)")
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit, color='orange')
plt.title("Company Profits - Horizontal Bar")
plt.xlabel("Profit ($)")
plt.show()
```

# Solution 5

A box plot (or box-and-whisker plot) displays the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is useful for identifying outliers and comparing distributions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box Plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
